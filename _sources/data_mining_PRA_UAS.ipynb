### Penjelasan Algoritma Local Outlier Factor (LOF) dan Alasan Pemilihannya

**Local Outlier Factor (LOF)** adalah algoritma deteksi outlier yang berbasis pada kepadatan (density-based). Algoritma ini mengukur seberapa terisolasi atau abnormal suatu titik data relatif terhadap tetangga terdekatnya, bukan relatif terhadap seluruh dataset secara global.

#### Konsep Utama

1.  **K-distance:** Jarak dari suatu titik data $p$ ke tetangga terdekat ke-$k$.
2.  **Reachability Distance:** Jarak yang "diperhalus" antara dua titik. Ini adalah jarak maksimum antara titik $p$ dan titik lain $o$, atau k-distance dari $o$, mana yang lebih besar. Ini membantu mengurangi fluktuasi statistik pada titik-titik yang sangat dekat.
3.  **Local Reachability Density (LRD):** Kebalikan dari rata-rata reachability distance dari suatu titik $p$ ke tetangga-tetangga k-nya. LRD mengukur seberapa padat area di sekitar titik $p$. Kepadatan yang lebih rendah berarti titik tersebut lebih terisolasi.

    $$LRD(p) = \frac{1}{\text{average}_{o \in N_k(p)} \text{reachability-distance}_k(p, o)}$$
    Di mana $N_k(p)$ adalah k-tetangga terdekat dari titik $p$.

4.  **Local Outlier Factor (LOF):** Rasio rata-rata LRD dari tetangga-tetangga k-nya terhadap LRD dari titik $p$ itu sendiri.

    $$LOF_k(p) = \text{average}_{o \in N_k(p)} \left( \frac{LRD(o)}{LRD(p)} \right)$$

    * Jika LRD titik $p$ mirip dengan LRD tetangganya, LOF akan mendekati 1. Ini menunjukkan titik tersebut berada di area dengan kepadatan yang mirip dengan tetangganya (bukan outlier).
    * Jika LRD titik $p$ secara signifikan lebih rendah dari LRD tetangganya, LOF akan jauh lebih besar dari 1. Ini menunjukkan titik tersebut berada di area yang kurang padat dibandingkan tetangganya, menjadikannya kandidat outlier.

#### Cara Kerja

LOF menghitung skor LOF untuk setiap titik data. Skor ini merepresentasikan tingkat "keanehan" lokal titik tersebut. Skor LOF yang tinggi menunjukkan bahwa titik tersebut memiliki kepadatan lokal yang jauh lebih rendah dibandingkan tetangganya, sehingga dianggap sebagai outlier.

#### Alasan Memilih LOF untuk Dataset Ini

Kami memilih Local Outlier Factor (LOF) untuk mendeteksi outlier pada dataset beras yang sudah diskalakan berdasarkan beberapa pertimbangan:

1.  **Berbasis Kepadatan Lokal:** Dataset kita memiliki fitur-fitur numerik yang distribusinya mungkin bervariasi (seperti yang terlihat dari histogram). LOF sangat efektif dalam mendeteksi outlier yang mungkin berada di luar pola umum data, bahkan jika pola data memiliki kepadatan yang berbeda di berbagai area. Metode ini tidak hanya mencari titik yang jauh dari semua titik lain secara global, tetapi titik yang aneh *relatif terhadap lingkungannya*.
2.  **Tidak Memerlukan Asumsi Distribusi:** LOF adalah metode non-parametrik. Ini tidak mengasumsikan bahwa data mengikuti distribusi statistik tertentu (seperti normal), yang membuatnya fleksibel untuk berbagai jenis data.
3.  **Cocok untuk Data Multidimensional:** Dengan 7 fitur numerik, dataset kita bersifat multidimensional. LOF bekerja dengan baik dalam ruang berdimensi tinggi untuk mengukur isolasi titik data.
4.  **Memberikan Skor Outlier:** LOF memberikan skor numerik untuk setiap sampel, yang memungkinkan kita untuk mengurutkan sampel berdasarkan tingkat "keanehan" dan menetapkan ambang batas (seperti yang dilakukan oleh `fit_predict` di scikit-learn yang menggunakan ambang batas -1 untuk memisahkan outlier dari inlier) untuk mengidentifikasi outlier.

Meskipun ada metode lain, LOF memberikan keseimbangan yang baik antara kemampuan mendeteksi outlier dalam berbagai kepadatan data dan kemudahan interpretasi skornya dalam konteks lokalitas. Untuk dataset berukuran menengah seperti ini, LOF adalah pilihan yang efisien dan efektif.

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np
import pandas as pd # Make sure pandas is imported here if not globally available

# Apply LOF for outlier detection
# Use the scaled NumPy array X_scaled directly
lof = LocalOutlierFactor(n_neighbors=20, contamination='auto')

# fit_predict returns -1 for outliers and 1 for inliers
outlier_labels = lof.fit_predict(X) # Use X_scaled (NumPy array)

# The opposite_of_outlier_scores_ attribute provides the LOF scores.
lof_scores = lof.negative_outlier_factor_

# Create the X_scaled_df DataFrame *after* LOF calculation,
# including the original scaled data plus the LOF results.
# We need X_scaled (NumPy array) and original column names from X
X_scaled_df = pd.DataFrame(X, columns=X.columns) # Recreate DataFrame from scaled array

# Add the LOF scores and labels to the scaled DataFrame
X_scaled_df['lof_score'] = lof_scores
X_scaled_df['is_outlier'] = outlier_labels

# Display samples with the lowest LOF scores (most likely outliers)
print("Samples with lowest LOF scores (most likely outliers):")
display(X_scaled_df.sort_values(by='lof_score').head())

# Display the count of samples identified as outliers (-1) and inliers (1)
print("\nCount of outliers (-1) and inliers (1):")
print(X_scaled_df['is_outlier'].value_counts())
```

hasil dari lof menyatakan bahwa yang termasuk outlier ini sebanyak 34 data

setelah itu kita hapus data outliernya agar pelatiahan model semakin bagus

```python
# Remove outliers from the scaled features and encoded target
# Keep only the rows where 'is_outlier' is 1 (inliers)
X_cleaned_scaled = X_scaled_df[X_scaled_df['is_outlier'] == 1].drop(columns=['lof_score', 'is_outlier'])
y_cleaned_encoded = y[X_scaled_df['is_outlier'] == 1]

# Display the shape of the cleaned dataframes to confirm rows were removed
print("Shape of X_cleaned_scaled after removing outliers:", X_cleaned_scaled.shape)
print("Shape of y_cleaned_encoded after removing outliers:", y_cleaned_encoded.shape)

# Display the first few rows of the cleaned scaled features
print("\nCleaned scaled features (first 5 rows):")
display(X_cleaned_scaled.head())
```

### Penjelasan Algoritma Gaussian Naive Bayes

**Gaussian Naive Bayes** adalah salah satu varian dari algoritma klasifikasi Naive Bayes. Algoritma ini didasarkan pada **Teorema Bayes** dan memiliki asumsi "naive" bahwa semua fitur bersifat independen satu sama lain, **diberikan kelasnya**.

#### Konsep Utama

1.  **Teorema Bayes:** Algoritma ini menggunakan Teorema Bayes untuk menghitung probabilitas bahwa suatu sampel termasuk dalam kelas tertentu (`P(Kelas | Fitur)`), berdasarkan probabilitas awal kelas (`P(Kelas)`) dan probabilitas fitur diberikan kelas (`P(Fitur | Kelas)`).

    Rumus Teorema Bayes:
    `P(Kelas | Fitur) = [ P(Fitur | Kelas) * P(Kelas) ] / P(Fitur)`

    *   `P(Kelas | Fitur)` (Posterior Probability): Probabilitas bahwa sampel termasuk dalam kelas tertentu, diberikan nilai fitur-fiturnya. Ini yang ingin kita prediksi.
    *   `P(Fitur | Kelas)` (Likelihood): Probabilitas munculnya nilai fitur tertentu, diberikan bahwa sampel tersebut termasuk dalam kelas tertentu.
    *   `P(Kelas)` (Prior Probability): Probabilitas awal bahwa sampel termasuk dalam kelas tertentu (tanpa melihat fiturnya). Ini bisa dihitung dari frekuensi kelas dalam data training.
    *   `P(Fitur)` (Evidence): Probabilitas munculnya nilai fitur tertentu secara keseluruhan. Ini adalah konstanta untuk semua kelas, jadi seringkali diabaikan dalam perbandingan antar kelas.

2.  **Asumsi "Naive" (Independensi Fitur):** Ini adalah asumsi kunci (dan seringkali tidak realistis) dari Naive Bayes. Asumsi ini mengatakan bahwa nilai dari setiap fitur tidak bergantung pada nilai fitur lainnya, selama kita mengetahui kelasnya. Secara matematis, ini berarti:

    `P(Fitur1, Fitur2, ..., FiturN | Kelas) = P(Fitur1 | Kelas) * P(Fitur2 | Kelas) * ... * P(FiturN | Kelas)`

    Asumsi inilah yang membuat Naive Bayes "naive" tetapi juga membuatnya sangat cepat dan efisien, karena perhitungan probabilitas menjadi jauh lebih sederhana.

3.  **Gaussian:** Varian Gaussian Naive Bayes digunakan ketika fitur-fitur bersifat **kontinu**. Algoritma ini mengasumsikan bahwa nilai-nilai fitur untuk setiap kelas mengikuti **distribusi Gaussian (Normal)**. Untuk setiap kelas dan setiap fitur, model akan menghitung mean (rata-rata) dan standard deviation (standar deviasi) dari nilai fitur dalam data training. `P(Fitur | Kelas)` kemudian dihitung menggunakan fungsi kepadatan probabilitas (Probability Density Function - PDF) dari distribusi Gaussian.

#### Cara Kerja

1.  **Pelatihan:**
    *   Untuk setiap kelas dalam data training, hitung `P(Kelas)` (proporsi sampel dalam kelas tersebut).
    *   Untuk setiap kelas dan setiap fitur, hitung mean dan standard deviation dari nilai fitur sampel dalam kelas tersebut. Ini adalah parameter distribusi Gaussian untuk fitur tersebut dalam kelas tersebut.

2.  **Prediksi:**
    *   Untuk sampel baru yang tidak diketahui kelasnya, hitung `P(Kelas | Fitur)` untuk setiap kelas menggunakan Teorema Bayes dan asumsi independensi.
    *   Karena `P(Fitur)` sama untuk semua kelas, kita hanya perlu membandingkan `P(Fitur | Kelas) * P(Kelas)`.
    *   Model memprediksi kelas yang memberikan nilai `P(Kelas | Fitur)` tertinggi.

#### Kelebihan

*   Sederhana dan mudah diimplementasikan.
*   Sangat cepat untuk dilatih dan diprediksi, bahkan pada dataset besar.
*   Tidak sensitif terhadap skala fitur.
*   Bagus sebagai baseline model atau ketika data training terbatas.

#### Kekurangan

*   Asumsi independensi fitur seringkali tidak terpenuhi di dunia nyata, yang dapat mengurangi performa jika fitur-fitur sangat berkorelasi.
*   Estimasi probabilitas bisa kurang akurat karena asumsi independensi.

Meskipun memiliki asumsi yang kuat, Gaussian Naive Bayes seringkali berkinerja surprisingly baik dalam praktiknya, terutama pada masalah klasifikasi teks atau ketika fitur-fitur memang relatif independen.

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Split the cleaned and scaled data into training and testing sets
# Using a standard split of 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_cleaned_scaled, y_cleaned_encoded, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Initialize and train the Gaussian Naive Bayes model
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = gnb.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy as a percentage
print(f"\nAccuracy of the Gaussian Naive Bayes model: {accuracy:.2%}")
```

dari hasil pelatiahn model mendapatkan akurasi sebanyak 91.80% akurasi ini cukup baik untuk data ini

## Hasil Evaluasi Model Gaussian Naive Bayes

Setelah melatih model Gaussian Naive Bayes dan melakukan prediksi pada data testing, kita mengevaluasi performanya menggunakan Confusion Matrix dan Classification Report.

### Confusion Matrix

Confusion Matrix memberikan ringkasan visual tentang seberapa baik model dalam mengklasifikasikan setiap kelas. Hasilnya adalah:

```python
from sklearn.metrics import confusion_matrix, classification_report

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Display the confusion matrix
print("\nConfusion Matrix:")
print(conf_matrix)

# Calculate and display the classification report
# This includes Precision, Recall, F1-Score, and Support for each class
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

#### Interpretasi Hasil Specific dari Confusion Matrix dan Classification Report

Berdasarkan hasil eksekusi kode sebelumnya:

**Confusion Matrix:**

dari hasil runningnya Ini berarti:
- **True Negatives (TN = 296):** Terdapat 296 sampel yang sebenarnya adalah kelas 0 (Cammeo) dan diprediksi dengan benar oleh model sebagai kelas 0.
- **False Positives (FP = 43):** Terdapat 43 sampel yang sebenarnya adalah kelas 0 (Cammeo) tetapi diprediksi salah oleh model sebagai kelas 1 (Osmancik). Ini sering disebut sebagai kesalahan "Type I error".
- **False Negatives (FN = 33):** Terdapat 33 sampel yang sebenarnya adalah kelas 1 (Osmancik) tetapi diprediksi salah oleh model sebagai kelas 0 (Cammeo). Ini sering disebut sebagai kesalahan "Type II error".
- **True Positives (TP = 381):** Terdapat 381 sampel yang sebenarnya adalah kelas 1 (Osmancik) dan diprediksi dengan benar oleh model sebagai kelas 1.

Dari matriks ini, kita bisa melihat bahwa model memiliki jumlah prediksi benar (296 + 381 = 677) yang jauh lebih banyak daripada prediksi salah (43 + 33 = 76).
