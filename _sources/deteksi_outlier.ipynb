# üåü Deteksi Outlier dengan K-Nearest Neighbors (KNN)

## üîç Apa itu Deteksi Outlier?
Deteksi outlier adalah proses mengidentifikasi titik data yang berbeda secara signifikan dari mayoritas data lainnya. Outlier bisa muncul karena:
- **Kesalahan pengukuran** ‚ùå
- **Variasi alami** üåø
- **Fenomena khusus** ‚≠ê

Mengapa penting?
- **Meningkatkan akurasi model** ü§ñ
- **Mengungkap pola tersembunyi** üìà
- **Mendeteksi anomali atau fraud** üö®

## ü§î Mengapa Menggunakan K-Nearest Neighbors (KNN)?
KNN adalah algoritma non-parametrik yang menentukan outlier berdasarkan jarak ke tetangga terdekat.

**Cara kerjanya:**
1. Tentukan jumlah tetangga **k**.
2. Hitung jarak dari tiap titik data ke **k** tetangga terdekat.
3. Jika jarak terlalu jauh (di atas threshold tertentu), maka data tersebut dianggap **outlier** üö©.












# üåü Euclidean Distance pada KNN

## üìè Rumus Euclidean Distance

Euclidean Distance digunakan dalam algoritma K-Nearest Neighbors (KNN) untuk mengukur jarak antara dua titik data dalam ruang fitur. Rumusnya adalah:

$$
d(p, q) = \sqrt{(q_1 - p_1)^2 + (q_2 - p_2)^2 + \dots + (q_n - p_n)^2}
$$

**Keterangan:**
- üìç $p$: titik data pertama $(p_1, p_2, \dots, p_n)$
- üéØ $q$: titik data kedua $(q_1, q_2, \dots, q_n)$
- üìä $n$: jumlah fitur (dimensi)

## üç≠ Contoh Kasus

Kita memiliki dataset tentang rasa buah üç≠üçã dan kelasnya. Tujuannya adalah memprediksi kelas data baru.

| Buah  | Kadar Gula (Brix) | pH (Keasaman) | Kelas  |
|------|------------------|-------------|-------|

| Lemon | 6                | 2.3         | Asam   |
| Mangga| 14               | 4.0         | Manis  |
| Jeruk | 7                | 2.8         | Asam   |
| ? (Data Baru) | 10               | 3.0         | ?     |

Kita ingin menentukan apakah data baru (Kadar Gula: 10 Brix, pH: 3.0) termasuk kelas **Manis** atau **Asam**.

### ‚ú® Perhitungan Euclidean Distance

1. **Jarak ke Apel (12, 3.5):** üç≠

$$
d = \sqrt{(10 - 12)^2 + (3.0 - 3.5)^2}
$$

$$
= \sqrt{(-2)^2 + (-0.5)^2}
$$

$$
= \sqrt{4 + 0.25}
$$

$$
= \sqrt{4.25} \approx 2.06
$$

2. **Jarak ke Lemon (6, 2.3):** üçã

$$
d = \sqrt{(10 - 6)^2 + (3.0 - 2.3)^2}
$$

$$
= \sqrt{(4)^2 + (0.7)^2}
$$

$$
= \sqrt{16 + 0.49}
$$

$$
= \sqrt{16.49} \approx 4.06
$$

3. **Jarak ke Mangga (14, 4.0):** üç≠

$$
d = \sqrt{(10 - 14)^2 + (3.0 - 4.0)^2}
$$

$$
= \sqrt{(-4)^2 + (-1.0)^2}
$$

$$
= \sqrt{16 + 1.0}
$$

$$
= \sqrt{17} \approx 4.12
$$

4. **Jarak ke Jeruk (7, 2.8):** üçã

$$
d = \sqrt{(10 - 7)^2 + (3.0 - 2.8)^2}
$$

$$
= \sqrt{(3)^2 + (0.2)^2}
$$

$$
= \sqrt{9 + 0.04}
$$

$$
= \sqrt{9.04} \approx 3.01
$$

## üîç Kode untuk Mencari KNN (K-Nearest Neighbors)

````python
import pandas as pd
from tabulate import tabulate
k = 3  # jumlah tetangga yang kamu mau
hasil = []

for i in range(len(result_df)):
    jarak = []
    for j in range(len(result_df)):
        if i != j:  # supaya tidak membandingkan diri sendiri
            total = ((result_df.loc[i, 'petal length'] - result_df.loc[j, 'petal length'])**2 +
                     (result_df.loc[i, 'petal width'] - result_df.loc[j, 'petal width'])**2 +
                     (result_df.loc[i, 'sepal_length'] - result_df.loc[j, 'sepal_length'])**2 +
                     (result_df.loc[i, 'sepal_width'] - result_df.loc[j, 'sepal_width'])**2)**0.5
            jarak.append(total)

    jarak.sort()  # urutkan jarak
    avg_distance = sum(jarak[:k]) / k  # ambil k tetangga terdekat
    hasil.append(avg_distance)

result_df['avg_distance'] = hasil

# Deteksi outlier (misalnya, data dengan jarak rata-rata > threshold tertentu)
threshold = result_df['avg_distance'].quantile(0.9999)  # ambil threshold 99% persentil
outliers = result_df[result_df['avg_distance'] > threshold]
# Menampilkan semua baris hasil perhitungan KNN
print("Hasil perhitungan jarak rata-rata (KNN):")
print(tabulate(result_df[['id', 'petal length', 'petal width', 'sepal_length', 'sepal_width', 'avg_distance']],
               headers='keys', tablefmt='grid'))

````

# üìê Penjelasan Perhitungan Euclidean Distance pada KNN pada code

## ‚ú® Rumus Euclidean Distance

Pada kode ini, perhitungan jarak antara dua titik data $p$ dan $q$ dalam ruang fitur menggunakan rumus **Euclidean Distance**:

$$
d(p, q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + \dots + (p_n - q_n)^2}
$$

Di dalam kode Python, rumus ini diterapkan untuk menghitung jarak antara satu baris data dengan semua baris data lainnya:

$$
d(p, q) = \sqrt{(p_{\text{petal length}} - q_{\text{petal length}})^2 + (p_{\text{petal width}} - q_{\text{petal width}})^2 + (p_{\text{sepal _length}} - q_{\text{sepal_length}})^2 + (p_{\text{sepal_width}} - q_{\text{sepal_width}})^2}
$$

- $p$ dan $q$: dua baris data berbeda (titik data)
- **Fitur**:
  - **petal length** (panjang kelopak bunga)
  - **petal width** (lebar kelopak bunga)
  - **sepal_length** (panjang sepal/kelopak luar)
  - **sepal_width** (lebar sepal/kelopak luar)

## üìè Rata-rata Jarak K Tetangga Terdekat

Setelah menghitung jarak antara satu titik data dengan semua titik lainnya, langkah selanjutnya adalah mencari **rata-rata jarak dari K tetangga terdekat**:

$$
\text{avg_distance} = \frac{d_1 + d_2 + \dots + d_k}{k}
$$

dengan:
- $k$: jumlah tetangga yang diinginkan (dalam kode: `k = 3`)
- $d_1, d_2, \dots, d_k$: $K$ jarak terkecil setelah diurutkan

Di kode, ini dilakukan dengan:

```python
avg_distance = sum(jarak[:k]) / k

````python
# Memilih baris yang termasuk outlier
outlier_df = result_df[result_df['avg_distance'] > threshold]

# Visualisasi
import matplotlib.pyplot as plt

plt.scatter(result_df.index, result_df['avg_distance'], label='Data')
plt.scatter(outlier_df.index, outlier_df['avg_distance'], color='red', label='Outlier')
plt.legend()
plt.title('Deteksi Outlier dengan KNN')
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Menghitung Z-Score
mean_distance = result_df['avg_distance'].mean()
std_distance = result_df['avg_distance'].std()
result_df['z_score'] = (result_df['avg_distance'] - mean_distance) / std_distance

# Menentukan outlier (Z-Score > 3 atau < -3)
outlier_df = result_df[np.abs(result_df['z_score']) > 3]

# Visualisasi Z-Score Plot
plt.figure(figsize=(8, 5))
plt.scatter(result_df.index, result_df['z_score'], label='Data')
plt.scatter(outlier_df.index, outlier_df['z_score'], color='red', label='Outlier')
plt.axhline(y=3, color='orange', linestyle='dashed', label='Threshold +3')
plt.axhline(y=-3, color='orange', linestyle='dashed', label='Threshold -3')
plt.legend()
plt.title('Deteksi Outlier dengan Z-Score Plot')
plt.show()
````

## üéØ Kesimpulan Deteksi Outlier dengan KNN

Berdasarkan implementasi dan visualisasi yang telah dilakukan, kita bisa menarik beberapa kesimpulan penting terkait deteksi outlier menggunakan K-Nearest Neighbors (KNN):

1. **KNN sebagai Metode Outlier Detection**
   - KNN efektif untuk mendeteksi outlier dengan mengukur jarak rata-rata ke tetangga terdekat.
   - Titik data yang memiliki jarak lebih jauh dari mayoritas titik lainnya akan diidentifikasi sebagai outlier.

2. **Signifikansi Threshold**
   - Penentuan threshold (contohnya, kuantil tertentu seperti 99.99%) penting untuk membedakan data normal dan data outlier.
   - Semakin tinggi threshold, semakin ketat kriteria untuk mendeteksi outlier.

3. **Visualisasi Membantu Identifikasi**
   - Plot visual seperti scatter plot membantu memahami pola data dan mengamati titik-titik yang menyimpang.
   - Outlier biasanya terlihat terisolasi dan jauh dari kluster data utama.

4. **Keterbatasan KNN dalam Outlier Detection**
   - KNN cenderung kurang efisien jika dataset terlalu besar karena proses perhitungan jarak yang memakan waktu.
   - Sensitif terhadap nilai K ‚Äî K yang terlalu kecil bisa membuat banyak false positive, sementara K yang terlalu besar bisa menyamarkan outlier.

üåü **Kesimpulan Utama:** KNN menawarkan pendekatan intuitif untuk mendeteksi outlier, namun hasil terbaik diperoleh dengan kombinasi pemilihan K yang tepat, threshold yang sesuai, dan visualisasi yang informatif.
