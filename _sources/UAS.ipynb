## LOF

LOF bersifat **lokal**. Ia mengukur kepadatan (density) di sekitar sebuah titik dan membandingkannya dengan kepadatan rata-rata di sekitar tetangga-tetangganya. Sebuah titik dianggap outlier jika ia berada di area yang **jauh lebih tidak padat (less dense)** daripada area di mana tetangganya berada.

---

### Konsep Kunci dan Rumus Matematis

Untuk memahami rumus LOF, kita perlu membangunnya dari beberapa konsep dasar. Mari kita sebut titik yang sedang kita analisis sebagai titik **A**.

#### Parameter Penting: `k`
Sebelum memulai, kita harus memilih satu parameter, yaitu `k`, yang merupakan jumlah tetangga terdekat yang akan kita pertimbangkan.

#### 1. Jarak ke-k (k-distance)
Ini adalah konsep paling dasar.
* **Definisi:** `k-distance(A)` adalah jarak dari titik A ke tetangga terdekatnya yang ke-`k`.
* **Tujuan:** Ini mendefinisikan "radius" lingkungan lokal di sekitar titik A. Semua titik di dalam radius ini (termasuk si tetangga ke-k) disebut sebagai k-nearest neighbors dari A, atau kita notasikan sebagai $N_k(A)$.

#### 2. Jarak Ketercapaian (Reachability Distance)
Ini adalah konsep yang sedikit rumit namun krusial untuk membuat perhitungan menjadi stabil.
* **Definisi:** Jarak ketercapaian dari titik A ke titik B, dinotasikan `reachability-distance_k(A, B)`, adalah nilai yang lebih besar antara jarak asli `d(A, B)` dan `k-distance` dari titik B.
* **Rumus:**
    $$
    \text{reachability-distance}_k(A, B) = \max \{ k\text{-distance}(B), d(A, B) \}
    $$
* **Tujuan:** Untuk "meratakan" jarak. Jika titik A sangat dekat dengan B (lebih dekat dari radius lingkungan B), kita tetap menganggap jaraknya seolah-olah sejauh radius lingkungan B (`k-distance(B)`). Ini mencegah hasil yang tidak stabil jika titik-titik berada dalam satu klaster yang sangat padat.

#### 3. Kepadatan Ketercapaian Lokal (Local Reachability Density - lrd)
Ini adalah ukuran kepadatan di sekitar titik A.
* **Definisi:** `lrd_k(A)` adalah kebalikan (inverse) dari rata-rata jarak ketercapaian dari titik A ke semua tetangganya di $N_k(A)$.
* **Rumus:**
    $$
    \text{lrd}_k(A) = \frac{1}{\frac{\sum_{B \in N_k(A)} \text{reachability-distance}_k(A, B)}{|N_k(A)|}}
    $$
* **Intuisi:**
    * Jika rata-rata jarak ketercapaian ke tetangga-tetangganya kecil (artinya tetangganya sangat dekat), maka nilai `lrd` akan **tinggi** (area padat).
    * Jika rata-rata jarak ketercapaian besar (artinya tetangganya berjauhan), maka nilai `lrd` akan **rendah** (area renggang).

#### 4. Local Outlier Factor (LOF) - Rumus Akhir
Akhirnya, kita sampai pada skor LOF itu sendiri.
* **Definisi:** Skor `LOF_k(A)` adalah rasio antara rata-rata kepadatan ketercapaian lokal (`lrd`) dari tetangga-tetangga A dengan kepadatan ketercapaian lokal (`lrd`) dari titik A itu sendiri.
* **Rumus:**
    $$
    \text{LOF}_k(A) = \frac{\frac{\sum_{B \in N_k(A)} \text{lrd}_k(B)}{|N_k(A)|}}{\text{lrd}_k(A)}
    $$

---

### Interpretasi Nilai LOF

Rumus di atas menghasilkan sebuah skor. Cara menginterpretasikannya adalah sebagai berikut:

* **LOF ≈ 1:** Kepadatan di sekitar titik A (`lrd(A)`) hampir sama dengan kepadatan rata-rata tetangganya. Ini berarti A adalah **inlier** (bukan outlier) yang berada di dalam sebuah klaster dengan kepadatan seragam.
* **LOF < 1:** Kepadatan di sekitar titik A lebih tinggi daripada kepadatan rata-rata tetangganya. Ini berarti A berada di titik yang sangat padat (misalnya, di pusat sebuah klaster). Ini jelas **bukan outlier**.
* **LOF > 1:** Kepadatan di sekitar titik A lebih rendah daripada kepadatan rata-rata tetangganya. Ini menandakan bahwa A berada di area yang lebih renggang daripada lingkungannya. Ini adalah **kandidat kuat outlier**. Semakin besar nilainya (misal, 1.5, 2.0, atau lebih), semakin tinggi kemungkinan titik tersebut adalah outlier.

### Kelebihan dan Kekurangan LOF

#### Kelebihan:
* **Efektif pada Data dengan Kepadatan Bervariasi:** Ini adalah kekuatan utamanya. LOF bisa menemukan outlier yang mungkin terlewat oleh metode lain.
* **Tidak Perlu Asumsi Distribusi:** Tidak mengasumsikan data terdistribusi normal.
* **Memberikan Skor:** Menghasilkan skor anomali yang lebih informatif daripada sekadar label biner "outlier" / "bukan outlier".

#### Kekurangan:
* **Komputasi Mahal:** Membutuhkan perhitungan jarak yang intensif, sehingga bisa menjadi lambat pada dataset yang sangat besar (kompleksitas sekitar O(n²)).
* **Sensitif terhadap Parameter `k`:** Pemilihan nilai `k` dapat mempengaruhi hasilnya.
* **Kurang Efektif pada Data Berdimensi Tinggi:** Seperti banyak algoritma berbasis jarak, efektivitasnya menurun seiring bertambahnya jumlah fitur (dimensi) karena konsep "jarak" menjadi kurang bermakna (curse of dimensionality).

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import LocalOutlierFactor
from tabulate import tabulate


df = X.copy()
df['Selector'] = y['Selector']


print("Jumlah missing values sebelum imputasi:\n", df.isnull().sum())
df['A/G Ratio'].fillna(df['A/G Ratio'].median(), inplace=True)
print("\nJumlah missing values setelah imputasi:\n", df.isnull().sum())


le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender']) # Male -> 1, Female -> 0 (sesuai urutan abjad)
print("\nDataframe setelah encoding Gender:")
display(df.head())


X_processed = df.drop('Selector', axis=1)
y_processed = df['Selector']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_processed)

X_scaled = pd.DataFrame(X_scaled, columns=X_processed.columns, index=X_processed.index)

print("\nDataframe fitur setelah penskalaan:")
display(X_scaled.head())

k = 3
contamination = 0.2
threshold = 1.8

numeric_cols_for_lof = X_scaled.select_dtypes(include=np.number).columns.tolist()
if 'Gender' in numeric_cols_for_lof:
    numeric_cols_for_lof.remove('Gender')

X_scaled_numeric = X_scaled[numeric_cols_for_lof]

lof = LocalOutlierFactor(n_neighbors=k, contamination=contamination)


lof.fit(X_scaled_numeric)
outlier_scores = -lof.negative_outlier_factor_

is_outlier_status = np.where(outlier_scores > threshold, -1, 1)


X_scaled['lof_score'] = outlier_scores
X_scaled['is_outlier'] = is_outlier_status # Menggunakan status berdasarkan threshold


print("Samples dengan skor LOF tertinggi (paling mungkin outlier):")
display(X_scaled.sort_values(by='lof_score', ascending=False).head())

outlier_count = X_scaled[X_scaled['is_outlier'] == -1].shape[0]
inlier_count = X_scaled[X_scaled['is_outlier'] == 1].shape[0]

print(f"\nJumlah data yang terdeteksi sebagai outlier (-1) berdasarkan threshold {threshold}: {outlier_count}")
print(f"Jumlah data yang terdeteksi sebagai inlier (1) berdasarkan threshold {threshold}: {inlier_count}")
```

```python
# Filter data untuk menghapus outlier
X_cleaned_scaled = X_scaled[X_scaled['is_outlier'] == 1].drop(['lof_score', 'is_outlier'], axis=1)
y_cleaned_encoded = y_processed.loc[X_cleaned_scaled.index] # Ambil target yang sesuai dengan index data bersih

print("Jumlah data setelah menghapus outlier:")
print(f"Fitur (X_cleaned_scaled): {len(X_cleaned_scaled)}")
print(f"Target (y_cleaned_encoded): {len(y_cleaned_encoded)}")

print("\nData fitur setelah menghapus outlier dan penskalaan (beberapa baris pertama):")
display(X_cleaned_scaled.head())

print("\nData target setelah menghapus outlier (beberapa baris pertama):")
display(y_cleaned_encoded.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
import numpy as np
import pandas as pd

X_data = X_cleaned_scaled.copy()
y_data = y_cleaned_encoded.copy()



X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42, stratify=y_data)

print("Ukuran data setelah dibagi (sebelum transformasi):")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")



numeric_cols_for_transform = X_train.select_dtypes(include=np.number).columns.tolist()
if 'Gender' in numeric_cols_for_transform:
    numeric_cols_for_transform.remove('Gender')

pt = PowerTransformer(method='yeo-johnson')

X_train_transformed = X_train.copy()
X_train_transformed[numeric_cols_for_transform] = pt.fit_transform(X_train[numeric_cols_for_transform])


X_test_transformed = X_test.copy()
X_test_transformed[numeric_cols_for_transform] = pt.transform(X_test[numeric_cols_for_transform])

print("\nData fitur latih setelah transformasi distribusi (beberapa baris pertama):")
display(X_train_transformed.head())

print("\nData fitur uji setelah transformasi distribusi (beberapa baris pertama):")
display(X_test_transformed.head())

```

### Naive Bayes
adalah algoritma klasifikasi yang didasarkan pada Teorema Bayes dengan asumsi "naif" mengenai independensi antar fitur. Meskipun asumsinya sederhana (naif), algoritma ini seringkali bekerja dengan baik dalam praktiknya, terutama pada tugas klasifikasi teks dan data dengan banyak fitur.

Berikut adalah konsep-konsep utamanya:

Teorema Bayes: Teorema Bayes menjelaskan hubungan antara probabilitas bersyarat. Dalam konteks klasifikasi, kita ingin menghitung probabilitas sebuah data (X) termasuk dalam sebuah kelas tertentu (Y), yaitu $P(Y|X)$$P(Y|X)$. Teorema Bayes menyatakan:

$P(Y|X) = \frac{P(X|Y) * P(Y)}{P(X)}$$P(Y|X) = \frac{P(X|Y) * P(Y)}{P(X)}$

$P(Y|X)$$P(Y|X)$: Probabilitas Posterior - Probabilitas bahwa data X termasuk dalam kelas Y. Ini yang ingin kita prediksi.
$P(X|Y)$$P(X|Y)$: Probabilitas Likelihood - Probabilitas kemunculan data X jika diketahui data tersebut termasuk dalam kelas Y. Ini dihitung dari data training.
$P(Y)$$P(Y)$: Probabilitas Prior - Probabilitas awal suatu data termasuk dalam kelas Y, tanpa mengetahui fitur X. Ini dihitung dari proporsi kelas di data training.
$P(X)$$P(X)$: Probabilitas Bukti (Evidence) - Probabilitas kemunculan data X secara keseluruhan. Untuk tugas klasifikasi, nilai ini sama untuk semua kelas, sehingga seringkali diabaikan dalam perbandingan probabilitas antar kelas.
Asumsi "Naif" (Naïve Assumption): Di sinilah letak kesederhanaan (dan nama "Naive") algoritma ini. Naive Bayes mengasumsikan bahwa semua fitur bersifat independen satu sama lain, mengingat kelasnya. Artinya, nilai satu fitur tidak mempengaruhi nilai fitur lainnya jika kita sudah mengetahui kelas data tersebut.

Dengan asumsi ini, probabilitas likelihood $P(X|Y)$$P(X|Y)$ dapat dipecah menjadi perkalian probabilitas untuk setiap fitur secara individual:

$P(X|Y) = P(x_1|Y) * P(x_2|Y) * ... * P(x_n|Y)$$P(X|Y) = P(x_1|Y) * P(x_2|Y) * ... * P(x_n|Y)$

Dimana $x_1, x_2, ..., x_n$$x_1, x_2, ..., x_n$ adalah nilai dari fitur-fitur dalam data X.

Proses Klasifikasi: Untuk mengklasifikasikan sebuah data baru, Naive Bayes menghitung probabilitas posterior $P(Y|X)$$P(Y|X)$ untuk setiap kemungkinan kelas (Y). Kelas yang memiliki probabilitas posterior tertinggi akan menjadi prediksi model.

$Prediksi = \arg\max_{Y} P(Y|X) = \arg\max_{Y} P(X|Y) * P(Y)$$Prediksi = \arg\max_{Y} P(Y|X) = \arg\max_{Y} P(X|Y) * P(Y)$

(Mengesampingkan $P(X)$$P(X)$ karena konstan untuk semua kelas)

Dengan asumsi naif:

$Prediksi = \arg\max_{Y} P(Y) * P(x_1|Y) * P(x_2|Y) * ... * P(x_n|Y)$$Prediksi = \arg\max_{Y} P(Y) * P(x_1|Y) * P(x_2|Y) * ... * P(x_n|Y)$

Varian Naive Bayes: Ada beberapa varian Naive Bayes, tergantung pada asumsi distribusi probabilitas fitur:

Gaussian Naive Bayes: Digunakan untuk fitur numerik kontinu. Ia mengasumsikan bahwa probabilitas fitur mengikuti distribusi Gaussian (Normal) untuk setiap kelas. Kita menghitung rata-rata dan standar deviasi fitur untuk setiap kelas dari data training.
Multinomial Naive Bayes: Umum digunakan untuk data diskrit, seperti hitungan kata dalam klasifikasi teks.
Bernoulli Naive Bayes: Digunakan untuk fitur biner (hadir atau tidak hadir), juga umum dalam klasifikasi teks (misalnya, apakah sebuah kata muncul dalam dokumen).
Kelebihan Naive Bayes:

Cepat dan Efisien: Proses pelatihan dan prediksi sangat cepat.
Tidak Membutuhkan Banyak Data Training: Dapat bekerja dengan baik bahkan dengan jumlah data training yang relatif kecil.
Tidak Sensitif terhadap Fitur yang Tidak Relevan: Fitur yang tidak relevan cenderung memiliki probabilitas yang hampir sama di semua kelas, sehingga tidak terlalu mempengaruhi hasil.
Baik untuk Data Berdimensi Tinggi: Cukup efektif pada dataset dengan banyak fitur.
Kekurangan Naive Bayes:

Asumsi Independensi yang Kuat: Asumsi naif bahwa fitur independen jarang terpenuhi di dunia nyata. Jika fitur sangat berkorelasi, performa bisa menurun.
Masalah Zero Probability: Jika sebuah kategori fitur tidak pernah muncul di data training untuk kelas tertentu, probabilitas likelihood-nya akan nol, yang dapat membuat seluruh probabilitas posterior menjadi nol. Ini biasanya diatasi dengan teknik smoothing (misalnya, Laplace smoothing).
Estimasi Probabilitas Kurang Baik: Meskipun sering baik dalam klasifikasi (memprediksi kelas yang benar), estimasi probabilitas posterior yang diberikan (nilai $P(Y|X)$$P(Y|X)$) mungkin tidak terlalu akurat.

### Decision Tree (Pohon Keputusan)

adalah algoritma machine learning supervised yang dapat digunakan untuk tugas klasifikasi maupun regresi. Algoritma ini bekerja dengan memecah dataset menjadi subset-subset yang lebih kecil secara rekursif berdasarkan nilai fitur, sambil pada saat yang sama sebuah pohon keputusan terkait secara inkremental dibangun. Hasil akhirnya adalah sebuah struktur pohon yang mirip dengan diagram alir, di mana setiap "internal node" merepresentasikan pengujian pada sebuah atribut (fitur), setiap "cabang" merepresentasikan hasil dari pengujian tersebut, dan setiap "leaf node" (atau terminal node) merepresentasikan label kelas (dalam klasifikasi) atau nilai (dalam regresi).

Berikut konsep-konsep utamanya:

Struktur Pohon:

Root Node: Node paling atas yang merepresentasikan seluruh dataset.
Internal Node: Node yang merepresentasikan pengujian pada sebuah fitur. Data dibagi berdasarkan hasil pengujian ini.
Branches: Koneksi antara node yang merepresentasikan hasil dari pengujian fitur.
Leaf Node: Node terminal yang merepresentasikan label kelas akhir (dalam klasifikasi) atau nilai prediksi (dalam regresi).
Proses Pembuatan Pohon (Induksi): Algoritma Decision Tree membangun pohon secara top-down, dimulai dari root node. Pada setiap node, algoritma memilih fitur "terbaik" untuk memecah data (splitting) berdasarkan kriteria tertentu.

Kriteria Pemecahan (Splitting Criteria): Tujuan dari pemecahan adalah untuk membuat subset data di setiap cabang menjadi semurni mungkin (pure), artinya sebagian besar data di subset tersebut memiliki label kelas yang sama. Dua kriteria umum untuk mengukur "kemurnian" atau "ketidakmurnian" (impurity) node adalah:

Gini Impurity: Mengukur seberapa sering elemen yang dipilih secara acak dari subset akan salah diberi label jika diberi label sesuai dengan distribusi label di subset tersebut. Nilai Gini Impurity berkisar antara 0 (sangat murni) hingga 0.5 (paling tidak murni untuk masalah klasifikasi biner).
Entropy: Mengukur tingkat ketidakpastian atau kekacauan dalam subset. Nilai Entropy berkisar antara 0 (sangat murni) hingga 1 atau lebih (sangat tidak murni).
Algoritma memilih fitur yang menghasilkan penurunan ketidakmurnian terbesar setelah pemecahan. Penurunan ketidakmurnian ini disebut Information Gain (jika menggunakan Entropy) atau penurunan Gini Impurity.

Kondisi Berhenti (Stopping Criteria): Proses pemecahan berhenti ketika salah satu kondisi berikut terpenuhi:

Semua data di node memiliki label kelas yang sama (node murni).
Tidak ada fitur yang tersisa untuk dipecah.
Kedalaman pohon mencapai batas maksimum yang ditentukan.
Jumlah sampel di node kurang dari batas minimum yang ditentukan.
Prediksi: Untuk mengklasifikasikan data baru, data tersebut dilewatkan dari root node ke bawah, mengikuti cabang sesuai dengan nilai fiturnya pada setiap node internal, hingga mencapai leaf node. Label kelas pada leaf node tersebut menjadi prediksi untuk data baru.

Kelebihan Decision Tree:

Mudah Diinterpretasikan: Struktur pohonnya mudah dipahami dan divisualisasikan, sehingga proses pengambilan keputusan model transparan.
Tidak Membutuhkan Penskalaan Fitur: Decision Tree tidak sensitif terhadap skala fitur.
Dapat Menangani Data Numerik dan Kategorikal: Dapat bekerja langsung dengan kedua jenis data (meskipun data kategorikal seringkali perlu di-encode).
Dapat Menangkap Hubungan Non-Linier: Mampu memodelkan hubungan yang kompleks dan non-linier antar fitur.
Kekurangan Decision Tree:

Cenderung Overfitting: Pohon yang terlalu dalam bisa sangat spesifik pada data training dan berkinerja buruk pada data baru. Teknik seperti pruning (memangkas cabang) atau membatasi kedalaman pohon diperlukan.
Sensitif terhadap Variasi Data Kecil: Perubahan kecil pada data training bisa menghasilkan pohon yang sangat berbeda.
Bias terhadap Fitur dengan Banyak Kategori: Jika ada fitur kategorikal dengan banyak nilai unik, algoritma cenderung memilih fitur tersebut.
Tidak Selalu Menghasilkan Pohon Optimal Global: Algoritma membangun pohon secara greedy (memilih pemecahan terbaik pada setiap langkah), yang tidak menjamin hasil pohon optimal secara global.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


X_train_data = X_train_transformed.copy()
y_train_data = y_train.copy()
X_test_data = X_test_transformed.copy()
y_test_data = y_test.copy()


model_dt = DecisionTreeClassifier(random_state=42)

model_dt.fit(X_train_data, y_train_data)


y_pred_dt = model_dt.predict(X_test_data)

accuracy_dt = accuracy_score(y_test_data, y_pred_dt)
print(f"Akurasi Model Decision Tree pada Data Uji: {accuracy_dt:.2%}")


report_dt = classification_report(y_test_data, y_pred_dt)
print("\nLaporan Klasifikasi Decision Tree:")
print(report_dt)


cm_dt = confusion_matrix(y_test_data, y_pred_dt)
print("\nConfusion Matrix Decision Tree:")
plt.figure(figsize=(6, 4))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix Decision Tree')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```

### K-Nearest Neighbors (KNN)
adalah algoritma supervised machine learning yang digunakan untuk klasifikasi dan regresi. Ini adalah algoritma yang sangat sederhana dan "malas" (lazy learner), artinya ia tidak benar-benar mempelajari model selama fase pelatihan, tetapi hanya mengingat dataset pelatihan. Seluruh pekerjaan dilakukan selama fase prediksi.

Berikut konsep-konsep utamanya:

Prinsip Dasar: Ide utama di balik KNN adalah bahwa titik data baru akan diklasifikasikan berdasarkan mayoritas kelas dari k tetangga terdekatnya di dataset pelatihan. "Tetangga terdekat" ditentukan berdasarkan jarak antar titik data.

Parameter k: Ini adalah parameter kunci dari algoritma KNN. k adalah jumlah tetangga terdekat yang akan dipertimbangkan saat membuat prediksi untuk titik data baru. Pemilihan nilai k sangat penting dan dapat mempengaruhi performa model.

Pengukuran Jarak: KNN menggunakan fungsi jarak untuk menentukan "kedekatan" antar titik data. Fungsi jarak yang paling umum digunakan adalah Jarak Euclidean, tetapi jarak lain seperti Jarak Manhattan juga bisa digunakan.

Jarak Euclidean: Jarak garis lurus antar dua titik dalam ruang multidimensional. $$d(p, q) = \sqrt{\sum_{i=1}^n (q_i - p_i)^2}$$$$d(p, q) = \sqrt{\sum_{i=1}^n (q_i - p_i)^2}$$
Jarak Manhattan: Jumlah jarak absolut di sepanjang semua sumbu. $$d(p, q) = \sum_{i=1}^n |q_i - p_i|$$$$d(p, q) = \sum_{i=1}^n |q_i - p_i|$$
Proses Pelatihan (Training): Sangat sederhana! Algoritma hanya menyimpan dataset pelatihan. Tidak ada model matematis yang kompleks dibangun selama pelatihan.

Proses Prediksi (Prediction) untuk Titik Data Baru: Ketika titik data baru tiba, KNN melakukan langkah-langkah berikut:

Hitung jarak antara titik data baru dengan setiap titik data di dataset pelatihan.
Pilih k titik data dari dataset pelatihan yang memiliki jarak terkecil ke titik data baru. Ini adalah "tetangga terdekat"nya.
Untuk Klasifikasi: Hitung jumlah titik data di antara k tetangga tersebut yang termasuk dalam setiap kelas. Titik data baru akan diberi label kelas yang paling sering muncul di antara k tetangga tersebut (mayoritas suara).
Untuk Regresi: Hitung rata-rata nilai target dari k tetangga tersebut. Nilai rata-rata ini akan menjadi prediksi untuk titik data baru.
Kelebihan KNN:

Sederhana dan Mudah Diimplementasikan: Konsepnya intuitif dan mudah dipahami.
Tidak Ada Asumsi Distribusi: Tidak mengasumsikan distribusi data (non-parametrik).
Fleksibel: Dapat digunakan untuk tugas klasifikasi dan regresi.
Efektif untuk Dataset Kecil-Menengah: Dapat bekerja dengan baik pada dataset yang ukurannya tidak terlalu besar.
Kekurangan KNN:

Komputasi Mahal saat Prediksi: Karena harus menghitung jarak ke semua titik data pelatihan untuk setiap prediksi baru, KNN bisa sangat lambat pada dataset besar.
Sensitif terhadap Skala Fitur: KNN sangat dipengaruhi oleh skala fitur karena berbasis jarak. Fitur dengan rentang nilai besar akan mendominasi perhitungan jarak. Penskalaan fitur (seperti Standard Scaling atau Min-Max Scaling) sangat penting sebelum menggunakan KNN.
Sensitif terhadap Outlier: Outlier dapat menarik batas keputusan dan mempengaruhi tetangga terdekat, sehingga mempengaruhi prediksi.
Pemilihan Nilai k: Memilih nilai k yang optimal bisa menjadi tantangan dan seringkali memerlukan eksperimen (misalnya, menggunakan cross-validation).
Tidak Bekerja Baik pada Data Berdimensi Tinggi: Pada data dengan banyak fitur (dimensi tinggi), konsep "jarak" menjadi kurang bermakna (curse of dimensionality), yang dapat menurunkan kinerja KNN.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

X_train_data = X_train_transformed.copy()
y_train_data = y_train.copy()
X_test_data = X_test_transformed.copy()
y_test_data = y_test.copy()

model_knn = KNeighborsClassifier()


model_knn.fit(X_train_data, y_train_data)

y_pred_knn = model_knn.predict(X_test_data)


accuracy_knn = accuracy_score(y_test_data, y_pred_knn)
print(f"Akurasi Model K-Nearest Neighbors pada Data Uji: {accuracy_knn:.2%}")


report_knn = classification_report(y_test_data, y_pred_knn)
print("\nLaporan Klasifikasi KNN:")
print(report_knn)


cm_knn = confusion_matrix(y_test_data, y_pred_knn)
print("\nConfusion Matrix KNN:")
plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix KNN')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```
